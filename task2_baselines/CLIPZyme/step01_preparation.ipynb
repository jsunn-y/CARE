Process the protein sequences and obtain AF structures

```python
swissprot = pd.read_csv('../processed_data/protein2EC_clustered50.csv')
uniprot_ids = swissprot['Entry'].unique()
swissprot
```

```python
#swissprot = swissprot[swissprot['Entry'].isin(uniprot_ids)]
swissprot['cif'] = [f"files/AF_structures/AF-{u}-F1-model_v4.cif" for u in swissprot['Entry'].values]
swissprot.rename(columns={'Entry': 'protein_id', 'Sequence': 'sequence'}, inplace=True)
swissprot
```

### process reactions

```python
def process_reaction(reaction):
    reactants, products = reaction.split('>>')
    reactants = reactants.split('.')
    products = products.split('.')

    reactants = [r for r in reactants if r != "[H+]"]
    products = [p for p in products if p != "[H+]"]

    reaction = "{}>>{}".format(".".join(reactants), ".".join(products))
    return reaction
```

```python
for split in ['easy', 'medium', 'hard']:
    reaction_df = pd.read_csv('../splits/task2/{}_reaction_test.csv'.format(split))
    # a lot of the mapped reactions are duplicated, that's a bit weird
    #reaction_df = reaction_df[reaction_df['Reaction'].isin(reactions)]
    reaction_df.drop_duplicates(subset='Mapped Reaction', inplace=True)
    reaction_df.rename(columns={'Mapped Reaction': 'reaction'}, inplace=True)
    
    reaction_df['reaction'] = reaction_df['reaction'].apply(process_reaction)

    #create filler columns for sequence, protein_id, and cif
    reaction_df['sequence'] = swissprot['sequence'].values[0]
    reaction_df['protein_id'] = swissprot['protein_id'].values[0]
    reaction_df['cif'] = swissprot['cif'].values[0]

    reaction_df.to_csv('reformatted/{}_reaction_test.csv'.format(split), index=False)
```

```python
#create filler columns for reaction
swissprot['reaction'] = reaction_df['reaction'].values[0]
swissprot.to_csv('reformatted/protein2EC_cluster50.csv', index=False)
```
