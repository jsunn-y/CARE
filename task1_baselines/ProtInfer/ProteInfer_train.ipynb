````python
import tensorflow as tf

import tqdm
import subprocess
import shlex
import tensorflow as tf
import glob

# Define the path to your TFRecord file
tfrecord_file = "/disk1/ariane/pycharm/CARE/proteinfer/testdata/train-00000-of-00001.tfrecord"

# Create a raw dataset from the TFRecord file
raw_dataset = tf.data.TFRecordDataset(tfrecord_file)
````

````python
import tensorflow as tf

# Make the records in the format that is expected by ProtInfer!
def create_tf_example(entry_dict):
    feature = {
        'id': tf.train.Feature(bytes_list=tf.train.BytesList(value=[entry_dict['sequence_id']])),
        'sequence': tf.train.Feature(bytes_list=tf.train.BytesList(value=[entry_dict['sequence']])),
        'label': tf.train.Feature(bytes_list=tf.train.BytesList(value=entry_dict['label'])),
    }
    return tf.train.Example(features=tf.train.Features(feature=feature))

def write_tfrecord_dataset(dataset, tfrecord_path):
    with tf.io.TFRecordWriter(tfrecord_path) as writer:
        for entry_dict in dataset:
            tf_example = create_tf_example(entry_dict)
            writer.write(tf_example.SerializeToString())

def write_vocab(ec_nums, filename):
    with open(filename, 'w+') as fout:
        fout.write(f'vocab_item\tvocab_index\n')
        for i, ec in enumerate(ec_nums):
            fout.write(f'{ec}\t{i}\n')
            
````

````python
# Read in our training data and see if we can convert to the right format and train it
import pandas as pd
train_df = pd.read_csv(f'../../CARE/splits/task1/protein_train.csv', index_col=0)


# now we just want to convert it into a dict with a seq_id, sequence and EC numbers as a list
grouped = train_df.groupby(['Entry', 'Sequence'])
train_values = []
for label, group in grouped:
    # Convert to bytes objects
    train_values.append({'sequence_id': str.encode(label[0]), 'sequence': str.encode(label[1]), 'label': [str.encode(f'EC:{ec}') for ec in group['EC number'].values]}) 
    
write_tfrecord_dataset(train_values, 'care_data/train.tfrecord')
````

````python
test_df = pd.read_csv(f'../../CARE/splits/task1/30-50_protein_test.csv', index_col=0)


# now we just want to convert it into a dict with a seq_id, sequence and EC numbers as a list
grouped = test_df.groupby(['Entry', 'Sequence'])
test_values = []
for label, group in grouped:
    # Convert to bytes objects
    test_values.append({'sequence_id': str.encode(label[0]), 'sequence': str.encode(label[1]), 'label': [str.encode(f'EC:{ec}') for ec in group['EC number'].values]}) 
    
write_tfrecord_dataset(test_values, 'care_data/30-50_protein_test.tfrecord')
````

````python
# Write the vocabulary and give each an index for all EC numbers in the train set
ec_nums = list(set(train_df['EC number'].values))
write_vocab(ec_nums, 'care_data/ec_vocab.tsv')
````
